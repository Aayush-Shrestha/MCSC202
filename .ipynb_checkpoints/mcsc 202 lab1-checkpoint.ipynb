```python
h0 = 1.2 # initial height in meters
v0 = 5.4 # initial velocity in m/s
g = 9.8 # acceleration due to gravity in m/s^2
t = 0.5 # time after being thrown in seconds
h = h0 + v0 * t - 0.5 * g * t**2
v = v0 - g * t
print("Height after {} seconds: {} meters".format(t, h))
print("Velocity after {} seconds: {} m/s".format(t, v))
t = 2.0 # time after being thrown in seconds
h = h0 + v0 * t - 0.5 * g * t**2
v = v0 - g * t
print("Height after {} seconds: {} meters".format(t, h))
print("Velocity after {} seconds: {} m/s".format(t, v))
```

```python
import numpy as np
# Create the array of 9 evenly spaced numbers from 0 to 29 (inclusive)
r = np.linspace(0, 29, 9)
# Find the square of each element of the array
r_squared = r ** 2
# Find twice the value of each element using addition
twice_r_addition = r + r
# Find twice the value of each element using multiplication
twice_r_multiplication = r * 2
# Print the results
print("Array r: {}".format(r))
print("Square of each element of r: {}".format(r_squared))
print("Twice the value of each element of r using addition: {}".format(twice_r_addition))
print("Twice the value of each element of r using multiplication: {}".format(twice_r_multiplication))
```

```python
# Constants
g = 9.8
h0 = 10
# Create the array of y values from 10 to 0 in increments of -0.5
y = np.arange(h0, -0.5, -0.5)
# Find the sequence of times when the ball passes each half meter using the equation t = sqrt(2*(h0-y)/g)
t = np.sqrt(2 * (h0 - y) / g)
# Find the average velocity for each time interval using the equation v = Δy/Δt = -0.5 / Δt
delta_y = -0.5 # Half meter increments
delta_t = np.diff(t) # Time intervals between each half meter increment
v = delta_y / delta_t
# Print the results
print("Sequence of times when the ball passes each half meter: {}".format(t))
print("Average velocity for each time interval: {}".format(v))
```

```python
a = float(input("Enter the value of coefficient a: "))
b = float(input("Enter the value of coefficient b: "))
c = float(input("Enter the value of coefficient c: "))
# Calculate the discriminant
discriminant = b**2 - 4*a*c
# Check the nature of the roots
if discriminant > 0:
    print("The roots are real and distinct.")
    root1 = (-b + math.sqrt(discriminant)) / (2*a)
    root2 = (-b - math.sqrt(discriminant)) / (2*a)
    print("Root 1 = ", root1)
    print("Root 2 = ", root2)
elif discriminant == 0:
    print("The roots are real and equal.")
    root1 = (-b) / (2*a)
    print("Root 1 = Root 2 = ", root1)
else:
    print("The roots are complex.")
    realPart = (-b) / (2*a)
    imaginaryPart = math.sqrt(abs(discriminant)) / (2*a)
    print("Root 1 = ", realPart, "+", imaginaryPart, "i")
    print("Root 2 = ", realPart, "-", imaginaryPart, "i")
```

```python
def factorial_while(n):
    if n == 0:
        return 1
    else:
        result = 1
        while n > 0:
            result *= n
            n -= 1
    return result
def factorial_for(n):
    if n == 0:
        return 1
    else:
        result = 1
        for i in range(1, n+1):
            result *= i
        return result
print(factorial_while(0)) # Expected output: 1
print(factorial_for(0)) # Expected output: 1
print(factorial_while(1)) # Expected output: 1
print(factorial_for(1)) # Expected output: 1
print(factorial_while(2)) # Expected output: 2
print(factorial_for(2)) # Expected output: 2
print(factorial_while(3)) # Expected output: 6
print(factorial_for(3)) # Expected output: 6
print(factorial_while(4)) # Expected output: 24
print(factorial_for(4)) # Expected output: 24
print(factorial_while(5)) # Expected output: 120
print(factorial_for(5)) # Expected output: 120
```

```python
import matplotlib.pyplot as plt
# Define the function
def f(x):
    return 3*x**2
# Generate x and y values
x = np.linspace(-3, 3, 100)
y = f(x)
# Create the plot
plt.plot(x, y)
# Add labels and title
plt.xlabel('x')
plt.ylabel('y')
plt.title('Plot of y = 3x^2')
plt.show()
```

```python
# Define x values
x = np.linspace(-2*np.pi, 2*np.pi, 1000)
# Define y values for sin(x) and cos(x)
y_sin = np.sin(x)
y_cos = np.cos(x)
# Plot the two functions on the same plot
plt.plot(x, y_sin, color='red', label='sin(x)')
plt.plot(x, y_cos, color='blue', label='cos(x)')
# Add a legend to the plot
plt.legend(loc='upper right')
# Show the plot
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D
# Define the curve
theta = np.linspace(0, 20*np.pi, 1000)
x = 2 * theta * np.cos(theta)
y = 2 * theta * np.sin(theta)
z = 1 + 2 * theta
# Create the 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot(x, y, z)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.show()
```

```python
def f(x, y):
    return (1 - x**2 + x**5 + y**5) * np.exp(-x**2 - y**2)
# Surface plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
x = np.linspace(-3, 3, 100)
y = np.linspace(-3, 3, 100)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)
ax.plot_surface(X, Y, Z, cmap='coolwarm')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.show()
# Contour plot
plt.figure()
plt.contourf(X, Y, Z, cmap='coolwarm')
plt.colorbar()
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```
