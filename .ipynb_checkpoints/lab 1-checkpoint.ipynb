Q1) Solution

```python
g = 9.8
h0 = 1.2
v0 = 5.4
```

```python
print('Height above the ground is ', h0 ,'m.')
print('And the initial velocity is ', v0 ,'m/s.')
```

```python
def answer(t):
    h = h0 + v0*t - 0.5*g*t*t
    v = v0-g*t
    h=round(h,2)
    v=round(v,2)
    print('After', t, 's the ball is at height',h,'m with velocity',v,'m/s.')
```

```python
answer(0.5)
```

```python
answer(2)
```

Q4) Solution

```python
g = 9.8
h0 = 10
y = np.arange(10,-0.1,-0.5)
```

```python
t = ((2*(h0-y))**0.5)/g
```

```python
z = np.zeros(21)
for i in range(1,len(t)):
    z[i] = -0.5/(t[i]-t[i-1])
z = np.round(z,3)
t = np.round(t,3)
```

```python
for i in range(0,len(t)):
    print('The ball is at height',y[i],'m at time',t[i],'s with avg velocity ',z[i],'m/s.')
```

Q5) Solution

```python
#this program finds nature of roots of ax^2 +bx + c = 0, a != 0
a , b , c = 1, 1 , 1 
d = (b**2) - (4*a*c)
```

```python
e1 = ((-b) + (d**0.5))/(2*a)
e1 = complex(round(e1.real,2),round(e1.imag,2))
e2 = ((-b) - (d**0.5))/(2*a)
e2 = complex(round(e2.real,2),round(e2.imag,2))
```

```python
if  d < 0:
    print('roots are distinct and imaginary')
    print('And the roots are ',e1.real,'+',e1.imag,'i ,',e2.real,'+',e2.imag,'i ')
elif d > 0:
    print('roots are distinct and real')
    print('And the roots are ',e1.real,e2.real)
else:
    print('roots are real and common')
    print('And the roots is ',e1.real)
```

Q13) Solution

```python
x = np.linspace(-3,3,100)
y = x
x, y = np.meshgrid(x, y)
```

```python
z = ( 1 - 0.5 * x + x**5 + y**5)*(np.e**((x*x)+(y*y)))
```

```python
#contour plot
plt.contourf(x, y, z);
plt.show()
```

```python
#Surface plot
surf = plt.axes(projection ='3d')
surf.plot_surface(x,y,z)
plt.show()
```
